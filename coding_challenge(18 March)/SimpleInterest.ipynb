```python
def SimpleInterest(p,t,r):
    print((p*t*r) / 100)
a = int(input("Enter the principal amount: "))
b = int(input("Enter the time period: "))
c = int(input("Enter the rate of interest: "))
SimpleInterest(a,b,c)
```
