```python
u = int(input("Enter units: "))
if u <= 100:
    bill = u * 1.5
elif u <= 200:
    bill = 100*1.5 + (u-100)*2.5
else:
    bill = 100*1.5 + 100*2.5 + (u-200)*4
print("Bill =", bill)
```
