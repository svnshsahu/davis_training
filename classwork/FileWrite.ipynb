```python
# Open file in write mode
f = open("sentences.txt", "w")

# Take 5 sentences from user
for i in range(5):
    sentence = input(f"Enter sentence: ")
    f.write(sentence)
print("--------------------------------")
# Close the file
f.close()

print("5 sentences have been written to the file.")
```
