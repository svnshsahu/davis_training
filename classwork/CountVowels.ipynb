```python
f = open("VowelsCount.txt")
if(f):
    data = f.read()

    vowels = "aeiouAEIOU"

    count = 0

    for ch in data:
        if ch in vowels:
            count += 1

    print("Number of vowels:", count)

else:
    print("Unable to get the file")
```
