```python
sentences = []

for i in range(20):
    s = input("Enter sentence : ")
    sentences.append(s + "\n")

f = open("sentences3.txt", "w")   # open file
f.writelines(sentences)           # write all sentences
f.close()                         # close file manually

print("Done!")
```
