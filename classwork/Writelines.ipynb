```python
# Create a list to store sentences
sentences = []

# Take 5 sentences from user
for i in range(5):
    s = input(f"Enter sentence : ")
    sentences.append(s + "\n")  # add newline manually

# Write all sentences at once
with open("sentences.txt", "w") as f:
    f.writelines(sentences)

print("Sentences written successfully.")
```
