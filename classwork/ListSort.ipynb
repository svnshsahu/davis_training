```python
#taking list of 10 numbers
numbers = [10,55,99,40,2,5,98,66,7,50]

#comparison using 2 loops
for i in range(0,10):
  for j in range(i+1,10):

    #swapping if condition satisfy
    if numbers[i]<numbers[j]:
      temp = numbers[i]
      numbers[i]=numbers[j]
      numbers[j]=temp

#print list
print(numbers)
```
